## Problem 2. Train a machine learning model using the training file to predict the values of the ‘tat_in_days’ column for the test file. ( Evaluation on the basis on RMSE score)

 

```python
import pandas as pd
from sklearn import metrics
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset_2=pd.read_csv('problem_2_train.csv')
```

```python
dataset_2.shape
```

```python
sns.heatmap(dataset_2.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
dataset_2.isnull().sum()
```

```python
dataset_2['mobmake'].value_counts()
```

```python
dataset_2.head()
```

```python
dataset_2['memcreateddate']=pd.to_datetime(dataset_2.memcreateddate)
dataset_2['mobilepurchasedate']=pd.to_datetime(dataset_2.mobilepurchasedate)
dataset_2['claimintdate']=pd.to_datetime(dataset_2.claimintdate)
dataset_2['pickupStartdate']=pd.to_datetime(dataset_2.pickupStartdate)
```

```python
dataset_2.dtypes
```

```python
dataset_2['pick_duration']=dataset_2['pickupStartdate']-dataset_2['claimintdate']
dataset_2['pick_duration']=dataset_2['pick_duration'].dt.days
dataset_2['claim_days']=dataset_2['claimintdate']-dataset_2['memcreateddate']
dataset_2['claim_days']=dataset_2['claim_days'].dt.days
```

```python
#dataset_2['delivery_time']=dataset_2['tat_in_days']-dataset_2['pick_duration']
```

```python
#sorted(dataset_2['hubid'].unique())
```

```python
dataset_2.head()
```

```python
dataset_2.columns
```

```python
X=dataset_2[['srno','mobmake', 'hubid', 'custpincodecategory','claimtype', 'custpincode', 'servicecntrid','pick_duration','claim_days']]
y=dataset_2['tat_in_days']
```

```python
#X.drop(['srno','hubpincode','memcreateddate','mobilepurchasedate','claimintdate','pickupStartdate'],axis=1,inplace=True)
```

```python
X.head()
```

```python
#mobmake = pd.get_dummies(X['mobmake'], prefix = 'mobmake', drop_first=True)
```

```python
custpincodecategory = pd.get_dummies(X['custpincodecategory'], prefix = 'custpincodecategory', drop_first=True)
```

```python
claimtype = pd.get_dummies(X['claimtype'], prefix = 'claimtype', drop_first=True)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()

X.iloc[:, 1] = labelencoder_X.fit_transform(X.iloc[:,1])

```

```python
#pickuppartnerid = pd.get_dummies(X['pickuppartnerid'], prefix = 'pickuppartnerid', drop_first=True)
```

```python
#deliverypartnerid = pd.get_dummies(X['deliverypartnerid'], prefix = 'deliverypartnerid', drop_first=True)
```

```python
#servicecntrid = pd.get_dummies(X['servicecntrid'], prefix = 'servicecntrid', drop_first=True)
```

```python
X.drop(['custpincodecategory','claimtype'],axis=1,inplace=True)
```

```python
X=pd.concat([X,custpincodecategory,claimtype],axis=1)
```

```python
X.shape
```

```python
X.head()
```

```python
X.corr()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=1)
```

```python
#linearity_check_df = pd.concat([pd.DataFrame(X),y],axis=1)
```

```python
#from sklearn.preprocessing import StandardScaler
#sc = StandardScaler()
#X_train.iloc[:,1:2] = sc.fit_transform(X_train.iloc[:,1:2])
#X_test.iloc[:,1:2] = sc.transform(X_test.iloc[:, 1:2])
```

###### Linear Regression

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(X_test)
```

```python
metrics.mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,y_pred))
```

```python
#sns.pairplot(X,kind='reg')
```

```python
regressor.score(X_test,y_test)
```

```python
print(regressor.coef_)
```

###### Decision Tree For Regression

```python
from sklearn.tree import DecisionTreeRegressor

clf = DecisionTreeRegressor(max_depth=4)

# Train Decision Tree Regressor
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
metrics.mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,y_pred))
```

```python
clf.score(X_test,y_test)
```
